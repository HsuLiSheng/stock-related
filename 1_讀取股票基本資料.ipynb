```python
class all_stock:
  page = "https://isin.twse.com.tw/isin/class_main.jsp?owncode=&stockname=&isincode=&market=1&issuetype=1&industry_code=&Page=1&chklike=Y"
  path = '/content/drive/Shareddrives/python 股票交易/stock_all'

  def __init_(self):
    pass

  def classification(df):
    classes = df.loc[:,'產業別']
    classes1 = classes.tolist()
    class_name = []
    for i in classes1:
      if i not in class_name:
        class_name.append(i)
    return class_name 

  def extract_stock(list1,classes):
    name = list1.loc[:,"產業別"] == classes
    name1 = name.tolist()
    df = pd.DataFrame(columns = ["有價證券代號","有價證券名稱","市場別","產業別","公開發行/上市(櫃)/發行日"])
    for i in range(len(name1)):
      if name1[i]:
        df = df.append(list1.iloc[i,:], ignore_index=True) ### df.append前面要有df存資料
    return df    
```

```python
stock = all_stock
path = stock.path
df = pd.read_excel(path+'/stock.xlsx')
class_name = stock.classification(df)
class_name
```

```python
name = '航運業'
df1 = stock.extract_stock(df,name)
df1
```
