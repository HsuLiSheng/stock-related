```python
def fundamental(data,index):
  try:
    if index == '殖利率(%)':
      ind = 2
    elif index == '本益比':
      ind = 4
    elif index == '股價淨值比': 
      ind = 5

    if len(data) > 30:
      rank = sorted(data, key=lambda x: float(x[ind].replace(',','')), reverse=True)[:25]
    else:
      rank = sorted(data, key=lambda x: float(x[ind].replace(',','')), reverse=True)
    return rank
    
  except:
    print('非有效基本面查詢')
    return    
```

```python
###
###每日成交量前20名證卷
###

def day_trade(date):
#date = '20220816'
    print('查詢日期:'+ date)
    html2 =urlopen ('https://www.twse.com.tw/exchangeReport/MI_INDEX20?response=json&date='+ date)
    content2 = html2.read().decode('utf-8')
    jcontent2 = json.loads(content2) ###利用json把數據整理更好看

    #print(jcontent2['fields'])
    chinese = jcontent2['fields'][0:6]
    chinese.append(jcontent2['fields'][8])
    chinese.append(jcontent2['fields'][10])
    print(chinese)

    data2 = jcontent2['data']
    data2_0 = [ i[:6] for i in data2]

    for i in range(len(data2)):
      data2_0[i].append(data2[i][8])
      if len(data2[i][9]) < 5:
        data2_0[i].append(data2[i][10])
      elif data2[i][9][19] == 'r':
        data2_0[i].append('+'+data2[i][10])
      else:
        data2_0[i].append('-'+data2[i][10])  

    return data2_0
#data2_0
```

```python
date = datetime.date.today()
try:
  if date.weekday()>=4:
    if date.weekday()==4:
      mon = date - datetime.timedelta(days=4)
      monday = datetime.date.strftime(mon,"%Y%m%d")
      fri = datetime.date.strftime(date,"%Y%m%d")
      print(monday+'~'+fri)
    else:
      mon = date - datetime.timedelta(days=date.weekday())
      monday = datetime.date.strftime(mon,"%Y%m%d")
      date = date - datetime.timedelta(days=date.weekday()-4)
      fri = datetime.date.strftime(date,"%Y%m%d")
      print(monday+'~'+fri)  
  
 
    day = 1
    most_stock = {}
    while(day<=5):
      date = datetime.date.strftime(mon,"%Y%m%d")
      data = day_trade(date)
      print(data[:5])
      for i in range(len(data)):
        if data[i][2] in most_stock.keys():
          most_stock[data[i][2]] += 1
        else:
          most_stock[data[i][2]] = 1
            
      mon += datetime.timedelta(days=1)
      day += 1
  else:    
    date = datetime.date.strftime(date,"%Y%m%d")
    print('日期'+date)
    data = day_trade(date)
    display(data)   
except:
    if datetime.date.today().weekday() >=4:
      date = datetime.datetime.strptime(date,"%Y%m%d")
      newdate = date - datetime.timedelta(days=1)
      date = datetime.date.strftime(newdate,"%Y%m%d")
    print('wrong date!!')
```
